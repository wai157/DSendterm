## Correlation Matrix

```python
dataset_new = SmallDS.copy()
lbe = LabelEncoder()
dataset_new['Location'] = lbe.fit_transform(dataset_new['Location'])
sns.heatmap(data=dataset_new.corr(), annot=True)
```

From the correlation matrix, we can draw the following conclusions:

* Price and Area: There is a positive correlation between Price and Area. This indicates that as the Area increases, the Price tends to increase as well.

* Price and Ward: There is a weak negative correlation between Price and Ward. This suggests that there is a very slight tendency for the Price to change as the Ward value changes, but the correlation is too weak to make any significant conclusions.

* Area and Ward: There is a very weak positive correlation between Area and Ward. This implies that there is a slight tendency for the Area to change as the Ward value changes, but the correlation is very weak and not substantial.

Overall, based on these correlation values, it can be concluded that the relationship between Price, Area, and Ward is not very strong. The correlation between Price and Area is the most notable, indicating a modest positive relationship. However, the correlations between Price and Ward, as well as Area and Ward, are very weak, suggesting that they have minimal influence on each other.
