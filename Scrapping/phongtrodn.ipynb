```python
#Import Libraries:
import os
from time import sleep
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests
import pickle
```

```python
page = 1
house_infos = list()
while True:
    url = requests.get("https://phongtrodn.com/phong-tro-da-nang/page/" + str(page))
    page_soup = bs(url.text, 'html.parser')
    
    try:
        infos = page_soup.body.div.find_all(class_ = 'info-real')
    except: break
    
    for idx in range(len(infos)):
        house_info = dict()
        house_info['Location'] = infos[idx].find("span").strong.get_text()
        house_info['Price'] = infos[idx].div.find_all('strong')[0].get_text()[14:-13].
        house_info['Area'] = infos[idx].div.find_all('strong')[1].get_text()
        try:
            house_infos.append(house_info)
        except: continue
    with open('save_page.txt','w') as file:
        file.write(str(page))
        file.close()
    if (page % 10 == 0):
        with open('../Raw_Data/phongtrodn.pickle','wb') as handle:
            pickle.dump(house_infos, handle, protocol=pickle.HIGHEST_PROTOCOL)
    page+=1
if (house_infos):
    with open('../Raw_Data/phongtrodn.pickle','wb') as handle:
        pickle.dump(house_infos, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open("../Raw_Data/phongtrodn.pickle","rb") as handle:
    dataset = pickle.load(handle)
dataset = pd.DataFrame(data = dataset)
dataset
```

```python
dataset.to_csv('../Raw_Data/phongtrodn.csv')
```
